## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
df['subtotal'] = df['unit_price'] * df['qty']

df.head()
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['subtotal'] = df['unit_price'] * df['qty']

df[['unit_price', 'qty', 'subtotal']].head(2)
```

```python

df['total_weight'] = df['unit_weight'] * df['qty']

df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['total_weight'] = df['unit_weight'] * df['qty']

df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df['line_price'] = round((df['subtotal'] + df['shipping_price']) * 1.0925, 2)

df[['subtotal', 'shipping_price', 'line_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = round((df['unit_cost'] * df['qty']) + df['shipping_price'], 2)

df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = round(df['line_price'] - df['line_cost'], 2)

df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
for client in top_clients:
    print(f"Client {client} had a total order price of ${df.loc[df['client_id'] == client, 'line_price'].sum():,.2f}.")
```

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
for client in top_clients:
    print(f"{client}: ${df.loc[df['client_id'] == client, 'line_price'].sum():,.2f}.")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary = pd.DataFrame(columns=['client_id', 'qty', 'total_shipping', 'total_price', 'total_cost', 'total_profit'])


for client in top_clients:
    total_units = int(df.loc[df['client_id'] == client, 'qty'].sum())
    total_shipping = round(df.loc[df['client_id'] == client, 'shipping_price'].sum(), 2)
    total_price = round(df.loc[df['client_id'] == client, 'line_price'].sum(), 2)
    total_cost = round(df.loc[df['client_id'] == client, 'line_cost'].sum(), 2)
    total_profit = round(df.loc[df['client_id'] == client, 'line_profit'].sum(), 2)
    
    summary.loc[len(summary)] = [int(client), total_units, total_shipping, total_price, total_cost, total_profit]

summary = summary.sort_values('total_profit', ascending=False)
summary = summary.astype({'client_id': 'int64', 'qty': 'int64', 'total_shipping': 'string', 'total_price': 'string', 'total_cost': 'string', 'total_profit': 'string'})

summary.head()
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary = summary.astype({'client_id': 'int64', 'qty': 'int64', 'total_shipping': 'string', 'total_price': 'string', 'total_cost': 'string', 'total_profit': 'string'})

summary['total_shipping'] = summary['total_shipping'].apply(lambda x: f"${(float(x) / 1000000):0.2f}M")
summary['total_price'] = summary['total_price'].apply(lambda x: f"${(float(x) / 1000000):0.2f}M")
summary['total_cost'] = summary['total_cost'].apply(lambda x: f"${(float(x) / 1000000):0.2f}M")
summary['total_profit'] = summary['total_profit'].apply(lambda x: f"${(float(x) / 1000000):0.2f}M")


summary = summary.rename(columns={'client_id': 'Client ID', 'qty': 'Total Units Purchased', 'total_shipping': 'Total Shipping Cost (Millions)', 'total_price': 'Total Price (Millions)', 'total_cost': 'Total Cost (Millions)', 'total_profit': 'Total Profit (Millions)'})

summary.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest

summary = summary.sort_values('Total Profit (Millions)', ascending=False)

summary.head()
```
